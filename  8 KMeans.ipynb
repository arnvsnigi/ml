```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

x,_=make_blobs(n_samples=200,centers=5,cluster_std=0.58,random_state=0)
x

```

```python
for k in range(2,6,1):
    model=KMeans(n_clusters=k)
    model.fit(x)
    pred=model.predict(x)
    plt.figure()
    plt.scatter(x[:,0],x[:,1],c=pred,s=50,alpha=0.75)
    plt.xlabel("X1")
    plt.ylabel("X2")
    plt.show()
```
