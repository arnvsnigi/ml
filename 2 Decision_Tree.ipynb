```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn import tree

df=pd.read_csv('Iris.csv')
df

# iris=load_iris()
# X=iris.data
# y=iris.target
# feature_names=iris.feature_names
# class_names=iris.target_names
```

```python
from sklearn.model_selection import train_test_split
X=df.drop(columns=['Id','Species'])
y=df['Species']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred=dtc.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

feature_namess = X.columns
class_namess = ['setosa', 'versicolor', 'virginica']
plt.figure(figsize=(20,10))  # Set the size of the plot
plot_tree(dtc, filled=True, feature_names=feature_namess, class_names=class_namess, rounded=True)
plt.show()
```
